Sacamos los IDS internos de The Fork de las ciudades principales de España

```python
def ids_ciudades_principales(headersList=headersList):
    reqUrl = "https://www.thefork.es/"
    payload = ""
    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)
    #Creamos una lista con los IDs de las principales ciudades en españa (city_list_codes_final)
    # Hacer la solicitud a la página web y obtener el HTML
    html = response.text
    # Parsear el HTML con BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    # Encontrar el JSON dentro del HTML (puedes usar la función find o select según la estructura del HTML)
    # Por ejemplo, si el JSON está dentro de una etiqueta <script> con un id específico, puedes hacer algo como esto:
    script_tag = soup.find('script', id='__NEXT_DATA__')
    json_str = script_tag.string
    # Una vez que tengas el JSON como una cadena de texto, puedes cargarlo como un objeto JSON
    json_data = json.loads(json_str)
    city_code_list=json_data["props"]["pageProps"]["initialApolloState"]["ROOT_QUERY"]['mainCities({"countryId":"60"})']
    city_list_codes_final=[]
    for i in city_code_list:
        city_list_codes_final.append(i['__ref'])
    dic_ciudades_principales={}
    for i in city_list_codes_final:
        city_data=json_data["props"]["pageProps"]["initialApolloState"][i]
        dic_ciudades_principales[city_data['id']]=city_data["slug"]
    return dic_ciudades_principales
```

Creamos una prueba para sacar de la primera ciudad todos sus restaurantes

```python
#Descargamos los datos de una URL concretao
def get_data_from_url(url, headersList=headersList):
    payload = ""
    response = requests.request("GET", url, data=payload,  headers=headersList)
    data_json=json.loads(response.text)
    return data_json


#Sacamos de una ciudad concreta los IDs de las ciudades cercanas
def get_ciudades_cercanas(ciudad_slug, ciudad_id, variable_sesion=variable_sesion):
    data_ciudad_concreta=get_data_from_url(f"https://www.thefork.com/_next/data/{variable_sesion}/en-US/search/cityTag/{ciudad_slug}/{ciudad_id}.json?p=1")
    dict_ciudades_cercanas={}
    for i in data_ciudad_concreta["pageProps"]["fallback"][f"/cities/{ciudad_id}/near?brandId=4&limit=15&radius=50000&sameCountryOnly=true"]:
        dict_ciudades_cercanas[i["id"]]=i["slug"]
    return dict_ciudades_cercanas


#Sacamos los IDs de los restaurantes de una ciudad y una única página concreta
def get_restaurantes_ciudad_page(ciudad_slug, ciudad_id, page,variable_sesion=variable_sesion):
    data_page= get_data_from_url(f"https://www.thefork.com/_next/data/{variable_sesion}/en-US/search/cityTag/{ciudad_slug}/{ciudad_id}.json?p={page}")
    ids_restaurantes={}
    restaurantes_ciudad=data_page["pageProps"]["searchPageResultsFetchResult"]["list"]
    for i in range(len(restaurantes_ciudad)):
        restaurant_id=restaurantes_ciudad[i]["restaurant"]["legacyId"]
        restaurant_slug=restaurantes_ciudad[i]["restaurant"]["slug"]
        ids_restaurantes[restaurant_id]=restaurant_slug
    return ids_restaurantes



#Sacamos los restaurantes de una ciudad concreta
def get_todos_restaurantes_ciudad(ciudad_slug, ciudad_id,variable_sesion=variable_sesion):
    data_page=get_data_from_url(f"https://www.thefork.com/_next/data/{variable_sesion}/en-US/search/cityTag/{ciudad_slug}/{ciudad_id}.json?p=1")
    numero_paginas=data_page["pageProps"]["searchPageResultsFetchResult"]["pagination"]["totalPage"]
    restaurantes_datos={}
    for i in range(1, numero_paginas+1):
        ids_restaurantes=get_restaurantes_ciudad_page(ciudad_slug, ciudad_id, i)
        restaurantes_datos.update(ids_restaurantes)

    return restaurantes_datos

def get_restaurantes_ciudad_y_alderedores(ciudad_slug, ciudad_id):
    ciudades_cercanas=get_ciudades_cercanas(ciudad_slug, ciudad_id)
    restaurantes_datos=get_todos_restaurantes_ciudad(ciudad_slug, ciudad_id)
    for i in ciudades_cercanas:
        restaurantes_datos.update(get_todos_restaurantes_ciudad(ciudades_cercanas[i], i))
    return restaurantes_datos

def get_ids_restaurantes_toda_espana():
    ciudades_principales_ids=ids_ciudades_principales()
    for i in ciudades_principales_ids:
        restaurantes_ciudad=get_restaurantes_ciudad_y_alderedores(ciudades_principales_ids[i], i)

        with open("data_all.json", "w") as file:
            json.dump(restaurantes_ciudad, file)
            print("listo")
    del restaurantes_ciudad
```

```python
def get_codigo_json_restaurante(slug, legacy_id):
    reqUrl = f"https://www.thefork.com/restaurant/{slug}-r{legacy_id}"
    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)
    html = response.text
    # Parsear el HTML con BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    # Encontrar el JSON dentro del HTML (puedes usar la función find o select según la estructura del HTML)
    # Por ejemplo, si el JSON está dentro de una etiqueta <script> con un id específico, puedes hacer algo como esto:
    script_tag = soup.find('script', id='__NEXT_DATA__')
    json_str = script_tag.string
    # Una vez que tengas el JSON como una cadena de texto, puedes cargarlo como un objeto JSON
    json_data = json.loads(json_str)
    return json_data


def data_restaurante(slug, legacy_id):
    json_data=get_codigo_json_restaurante(slug, legacy_id)
    dato_necesario_id_rest=json_data["props"]["pageProps"]["initialApolloState"]["ROOT_QUERY"]['restaurant({\"restaurantId":'+'"'+str(legacy_id)+'"'+'})']["__ref"]
    type_1=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["__typename"]
    try:
        city_id=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["city"]["__ref"]
        city_name=json_data["props"]["pageProps"]["initialApolloState"][city_id]["name"]
    except:
        city_id=None
        city_name=None
    try:
        michelin=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["guides"]["michelinV2"]
    except:
        michelin=None
    try:
        slug=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["slug"]
    except:
        slug=None
    try:
        bookable=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["isBookable"]
    except:
        bookable=None
    try:
        street=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["address"]["street"]
        zipCode=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["address"]["zipCode"]
        locality=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["address"]["locality"]
        country=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["address"]["country"]

        longitude=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["geolocation"]["longitude"]
        latitude=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["geolocation"]["latitude"]
    except:
        street=None
        zipCode=None
        locality=None
        country=None
        longitude=None
        latitude=None
    try:
        cantidad_metodos_pago=len(json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["paymentAccepted"])

        reservas_today=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["hasReservations"]["today"]
        reservas_last_week=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["hasReservations"]["lastWeek"]
        rate_Distinction=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["rateDistinction"]
        numero_fotos=len(json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["photos"])
        is_Affiliated=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["isAffiliated"]
        name=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["name"]
        average_Price=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["averagePrice"]
        numero_awards=len(json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["awards"])
    except:
        cantidad_metodos_pago=None
        reservas_today=None
        reservas_last_week=None
        rate_Distinction=None
        numero_fotos=None
        is_Affiliated=None
        name=None
        average_Price=None
        numero_awards=None
    #Valoraciones
    try:
        review_count_dato_nec=json_data["props"]["pageProps"]["initialApolloState"]["ROOT_QUERY"]['ratingSummary({\"restaurantId":'+'"'+str(legacy_id)+'"'+'})']["__ref"]
        review_count=json_data["props"]["pageProps"]["initialApolloState"][review_count_dato_nec]["reviewCount"]
        rating_count=json_data["props"]["pageProps"]["initialApolloState"][review_count_dato_nec]["ratingCount"]
        food_rating=json_data["props"]["pageProps"]["initialApolloState"][review_count_dato_nec]["food"]["average"]
        service_rating=json_data["props"]["pageProps"]["initialApolloState"][review_count_dato_nec]["service"]["average"]
        ambience_rating=json_data["props"]["pageProps"]["initialApolloState"][review_count_dato_nec]["ambience"]["average"]
    except:
        review_count=None
        rating_count=None
        food_rating=None
        service_rating=None
        ambience_rating=None
    #Tipo de comida
    try:
        tipo_comida=json_data["props"]["pageProps"]["initialApolloState"][dato_necesario_id_rest]["servesCuisine"]
    except:
        tipo_comida=None
    return type_1, city_name,michelin,slug,bookable,street,zipCode,locality,country,longitude,latitude,cantidad_metodos_pago,reservas_today,reservas_last_week,rate_Distinction,numero_fotos,is_Affiliated,name,average_Price,numero_awards,review_count,rating_count,food_rating,service_rating,ambience_rating,tipo_comida
```
